## **Title of Project** - **Mielage Prediction**

## Objective - The objective is to predict the mileage of a particular car in city driving, given data of some parameters for hundreds of cars.

Data Source: This dataset was taken from the StatLib library which is maintained at Carnegie Mellon University. The dataset was used in the 1983 American Statistical Association Exposition.

Attributes:

1. mpg:           continuous
2. cylinders:     multi-valued discrete
3. displacement:  continuous
4. horsepower:    continuous
5. weight:        continuous
6. acceleration:  continuous
7. model year:    multi-valued discrete
8. origin:        multi-valued discrete
9. car name:      string (unique for each instance)

```python
# Import Library
import pandas as pd
import sklearn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Read Data
mielage = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/MPG.csv')
```

```python
mielage.head()
```

```python
mielage.nunique()
```

```python
mielage.info()
```

```python
# Describe data
mielage.describe()
```

```python
mielage.corr()
```

```python
# Removing missing values
mielage = mielage.dropna()
```

```python
mielage.info()
```

```python
# Data visualisation
sns.pairplot(mielage, x_vars= ['displacement', 'horsepower', 'weight', 'acceleration', 'mpg'], y_vars=['mpg']);
```

```python
sns.regplot(x = 'displacement', y = 'mpg', data = mielage);
```

```python
# Data Pre-processing
mielage.isnull().sum()
mielage.info()
```

```python
# Define Target Variable (y) and Feature Variables (x)
mielage.columns
```

```python
y = mielage['mpg']
```

```python
x = mielage[['displacement', 'horsepower', 'weight',
       'acceleration']]
```

```python
x.shape
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Modeling
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# Prediction
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
# Model accuracy
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

Explanation: The mileage of a particular car in city driving, given data of some parameters for hundreds of cars has been predicted using Linear Regression analysis.
